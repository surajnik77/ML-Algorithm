```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load dataset from sklearn.datasets
df = pd.read_csv("titanic.csv")
df
```

```python
df = df[['Survived','Age','Sex']]
df = pd.get_dummies(df, columns=['Sex'])
df.dropna(inplace=True)
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
plt.scatter(df['Age'],df['Survived'])
```

```python
df.describe()
```

```python
X = df.drop('Survived',axis=1)
X[:2]
```

```python
y = df.Survived
y[:2]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred[:5]
```

```python
y_test[:5].values
```

```python
from  sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
model.score(X_test,y_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
f1_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
#Heatmap for check actual and predicted ooutput
sns.heatmap(cm,annot=True,xticklabels=['Dead','Survived'],yticklabels=['Dead','Survived'])
plt.xlabel('Predicted Result')
plt.ylabel('Actual Result')
plt.title('Heatmap for check actual and predicted ooutput')
```
