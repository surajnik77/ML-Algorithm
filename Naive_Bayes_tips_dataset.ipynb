```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import confusion_matrix,classification_report
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

**we have here categorical data in features so encoded this data.**

#Data Preprocessing

# here we have to predict time of dinner on the basis of other features

```python
df.time.unique()
```

```python
# OneHotEncoder-  Encoding the one column categorical data into seperart categorical column.
onehotencoder = OneHotEncoder(handle_unknown='ignore')
onehotencoder.fit(df[['sex']])
encoded = onehotencoder.transform(df[['sex']])
df[onehotencoder.categories_[0]] = encoded.toarray()
```

```python
df.head()
```

```python
#OrdinalEncoder - Encoding Data in order or rank form
oe = OrdinalEncoder()
df[['smoker','day']] = oe.fit_transform(df[['smoker','day']])
df.head()
```

```python
oe.categories_
```

```python
X = df.drop(['sex','time'],axis=1)
X[:5]
```

```python
y = df.time
y[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
le = LabelEncoder()
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
X_train[:5]
```

```python
y_train[:5]
```

```python
#GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
```

```python
y_pred = gnb.predict(X_test)
y_pred[:5]
```

```python
y_test[:5]
```

```python
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
bnb = BernoulliNB()
bnb.fit(X_train,y_train)
```

```python
y_pred1 = bnb.predict(X_test)
y_pred1[:5]
```

```python
print(confusion_matrix(y_pred1,y_test))
print(classification_report(y_pred1,y_test))
```

```python
mnb = MultinomialNB()
mnb.fit(X_train,y_train)
```

```python
y_pred2 = mnb.predict(X_test)
y_pred2[:5]
```

```python
print(confusion_matrix(y_pred2,y_test))
print(classification_report(y_pred2,y_test))
```

**here we select GaussianNB Model for final modeling with Accuracy 92%**
